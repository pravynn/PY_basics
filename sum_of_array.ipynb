```python
def main(input_list):
    ans = 0
    N = len(input_list)

    for i in range(N):
        ans += input_list[i]

    print(ans)

# Example Usage:
input_list1 = [5, 1, 2, 3, 4, 5]
input_list2 = [4, 10, 50, 40, 80]

# Call the main function with different input lists
main(input_list1[1:])  # Remove the first element (size) before passing to the function
main(input_list2[1:])  # Remove the first element (size) before passing to the function
```
