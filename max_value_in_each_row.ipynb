```python
def max_in_each_row(matrix):
    result = []

    for row in matrix:
        max_value = max(row)
        result.append(max_value)

    return result

# Example Input
input_matrix_1 = [[1, 2], [1, 3]]
input_matrix_2 = [[1, 2, 3]]

# Example Output
output_result_1 = max_in_each_row(input_matrix_1)
output_result_2 = max_in_each_row(input_matrix_2)

# Print the results
print("Output 1:", output_result_1)
print("Output 2:", output_result_2)
```
